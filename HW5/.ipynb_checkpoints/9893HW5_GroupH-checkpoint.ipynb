1. Carry out the SSA analysis of daily SP 500 along the lines discussed in the
notes for the stock of AAPL. You may want to use the Spider ETF SPY as

We implemented the SSA algorithm and verified the components obtained represent the behavior of the price series. 

```python
import pandas as pd
import matplotlib.pyplot as plt


# read data and plot it
spy = pd.read_csv('SPY.csv', usecols=['Date','Adj Close'], 
                 index_col=0, parse_dates=True,
                 infer_datetime_format=True)
spy = spy.dropna()
plt.figure()
plt.plot(spy,label='SPY')
plt.title('SPY daily adjusted closes')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```

```python
import scipy.linalg as linalg
import numpy as np

# Embedding steps
# First, form the trajectory matrix (Hankel matrix of lagged data)

def embed(data, l):
    n = len(data)
    k = n - l + 1
    # build hankel matrix with l rows and k columns
    chi = np.matrix(linalg.hankel(data, np.zeros(l))).T[:,:k]
    
    # calculate the elementary matrices and principal components 
    # from the trajectory matrix using its SVD decomposition
    r = np.linalg.matrix_rank(chi)
    S = chi*chi.T
    u, svalues, v = linalg.svd(S)
    U = np.matrix(u)
    V = np.matrix(v)
    sv = np.sqrt(svalues)
    V_i = {}
    pc_u = {}
    pc_v = {}
    Chi = {}
    for i in range(r):
        V_i[i] = ((1/sv[i])*chi.T)*U[:,i]
        Chi[i] = (sv[i]*U[:,i])*V_i[i].T
        pc_u[i] = np.matmul(chi.T, U[:,i])
        pc_v[i] = sv[i]*V_i[i]
    return [s, V, Chi, pc_u, pc_v]

# Reconstruction steps
# Partition into disjoint sets and sum 
# Perform diagonal averaging over the resulting sets to
# create the time series components
# Assume l < k for now, we expect the optimal rolling window to be reasonably
# smaller than the total number of samples for the original time series
def reconstruct(chi, rank):
    X_k = {}
    for i in range(rank):
        m_i = np.matrix(chi[i])
        X_k[i] = [np.mean(m_i[::-1, :].diagonal(i)) for i in 
           range(-m_i.shape[0] + 1, m_i.shape[1])]
    return X_k
```

First, we can calculate the cumulative weights of the eigenvectors found. For example, using a rolling window of 251 days (a year) for the SPY data, the first 5 components will contain approximately 87% of the information contained in the original samples.  
We experimented with different values of L as to try to incorporate most relevant features of the time series and a year both makes sense from the seasonality perspective and it seems to give us results that agree with our expectations. 

```python
# Choose L = the length of the rolling window and calculate 
# K, the number of lagged vectors, as T - L + 1 
L = 251
data = spy['Adj Close']
s, V, chi, pc_u, pc_v = embed(data, L)

# Calculate contribution of the first 5 SSA components
components = 5
cumulative = 0
for i in range(components):
    total = np.sum(s)
    cumulative += s[i] 
    print("CW%d = %f, W%d = %f"%(i+1,cumulative/total,i+1,s[i]/total))
```

Next, we reconstruct the components from the elementary matrices and observe the contributions of each to the sample data.

```python
x_k = reconstruct(chi,L)

plt.figure()
plt.plot(data,label='SPY')
plt.title('SPY adjusted closes')
plt.xlabel('Date')
plt.ylabel('Adj. Close')
plt.show()

plt.figure()
plt.plot(x_k[0],label='Comp1')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(x_k[1],label='Comp2')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(x_k[2],label='Comp3')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(x_k[3],label='Comp4')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(x_k[4],label='Comp5')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()
```

The first component indicates there is a strong upward trend component with a visible change of slope around mid-2016. Subsequent components seem to indicate behavior cycles of approximately 3 years (Component 2), 1 year (Components 3, 4) and 6 months (Component 5).

Let's check that adding each of the components does give us an incrementally better representation of the sample data

```python
from operator import add
C12 = list(map(add, x_k[0], x_k[1]))
C13 = list(map(add, C12, x_k[2]))
C14 = list(map(add, C13, x_k[3]))
C15 = list(map(add, C14, x_k[4]))

plt.figure()
plt.plot(data,label='SPY')
plt.title('SPY adjusted closes')
plt.xlabel('Date')
plt.ylabel('Adj. Close')
plt.show()

plt.figure()
plt.plot(x_k[0],label='Comp1')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(C12,label='Comp1+2')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(C13,label='Comp1+2+3')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(C14,label='Comp1+2+3+4')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()

plt.figure()
plt.plot(C15,label='Comp1+2+3+4+5')
plt.xlabel('Sample #')
plt.ylabel('Contribution to prices')
plt.legend()
plt.show()
```

2. Prove formula (43) in Lecture Notes #4

Shannon entropy of a normal random variable:
$$H(x) = - \int p(x)log(p(x))dx = \frac{1}{2} log \left(2\pi e \sigma^2\right)$$
where $p(x) = \frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x - \mu)^2}{2 \sigma^2}}$

\begin{align}
H(x) =& - \int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x - \mu)^2}{2 \sigma^2}} log \left(\frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x - \mu)^2}{2 \sigma^2}} \right)dx\\
=& - \int_{-\infty}^{\infty} \frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x - \mu)^2}{2 \sigma^2}} \left( -log \sqrt{2\pi \sigma^2} - \frac{(x - \mu)^2}{2 \sigma^2} \right)dx\\
=& \left( log \sqrt{2\pi \sigma^2} \right) \frac{1}{\sqrt{2\pi \sigma^2}} \int_{-\infty}^{\infty} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}dx   + \frac{1}{2 \sigma^2} \frac{1}{\sqrt{2\pi \sigma^2}}\int_{-\infty}^{\infty} (x - \mu)^2 e^{-\frac{(x - \mu)^2}{2 \sigma^2}}dx\\
=& log \sqrt{2\pi \sigma^2} \times 1   + \frac{1}{2 \sigma^2} E\left[(x-\mu)^2\right]\\
=& \frac{1}{2} log \left(2\pi \sigma^2\right) + \frac{1}{2 \sigma^2} \sigma^2\\
=& \frac{1}{2}\left( log \left(2\pi \sigma^2\right) + 1\right)\\
=& \frac{1}{2}\left( log \left(2\pi \sigma^2\right) + log(e) \right)\\
=& \frac{1}{2} log \left(2\pi e \sigma^2\right)\\
\end{align}
